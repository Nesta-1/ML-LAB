```python
# 1.	Write a program to calculate the sum of digits of a given number.

num = int(input("Enter a number: "))
s = 0
while num > 0:
    s += num % 10
    num //= 10
print("Sum of digits:", s)
```

```python
# 2.	Write a program to print a multiplication table up to 10 for a user-entered number.
n = int(input("Enter a number: "))
for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

```python
# 3.	Create a pie chart showing distribution of categorical data.
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Categorical Data Distribution")
plt.show()
```

```python
# 4.	Compare the performance of different data loading methods in pandas.

import pandas as pd
import time

file_path = r"C:\Nesta\KL\SEM 4\ML\ML LAB\week_1\data.csv"

# read_csv
start = time.time()
df_csv = pd.read_csv(file_path)
print("read_csv time:", time.time() - start)

# read_table
start = time.time()
df_table = pd.read_table(file_path, delimiter=",")
print("read_table time:", time.time() - start)

# read_csv with Python engine (slower, for comparison)
start = time.time()
df_csv_py = pd.read_csv(file_path, engine="python")
print("read_csv (python engine) time:", time.time() - start)

```
